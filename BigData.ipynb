```python
scz2 = pd.read_csv(path, sep='\t', chunksize=25000, low_memory=False, header=0, iterator=True)
```

```python
maxs = [chunk\
        .query('gene_symbol != "gene_symbol"')\
        .drop(DO_NOT_NEED + ['gene_id'], axis=1)\
        .set_index('gene_symbol')\
        .astype(float)\
        .groupby(level=0)\
        .agg(AGG) for chunk in scz2]
```
