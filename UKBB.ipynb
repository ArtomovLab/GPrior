```python
import os
import glob
import pandas as pd
```

```python
MAF = 0.05
pval = 0.0001

def cut_variants_with_little_MAF(df, MAF):
    filtered_df = df[df['minor_AF'] >= MAF]
    filtered_df.index = range(len(filtered_df.index))
    return filtered_df

def cut_variants_with_high_pval(df, pval):
    filtered_df = df[df['pval'] <= pval]
    filtered_df.index = range(len(filtered_df.index))
    return filtered_df
```
