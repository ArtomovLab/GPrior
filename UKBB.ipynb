```python
count = 0
for efo in efos.index[234:]:
    
    'https://www.ebi.ac.uk/gwas/rest/api/studies/search/findByDiseaseTrait?diseaseTrait=Asthma'
    
    EFO = efo
    trait = f'http://www.ebi.ac.uk/efo/{EFO}'
    request_url = f'https://www.ebi.ac.uk/gwas/rest/api/studies/search/findByEfoUri?uri={trait}'
    resp = requests.get(request_url)
    
    terms = list()
    for study in resp.json()['_embedded']['studies']:
        terms.append(study['diseaseTrait']['trait'])
    terms = set(terms)
    
    efos.loc[efo, 'terms'] = str(terms).replace('}', '').replace('{', '').replace("'", '')
    
    count += 1
    
    if count % 1000 == 0:
        print(count)
    
```
